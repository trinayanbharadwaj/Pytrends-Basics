```python
# Importing Pytrends
from pytrends.request import TrendReq

# we create a pytrends object to request data from google trends and specify the language.
pytrends = TrendReq(hl="en-US")

# We make a list of search terms.The limit is 5 keywords.
keywords = ["Python", "JavaScript", "Java", "Ruby", "Perl"]

# This is the command to extract data from google trends. Here, we specify our search terms and the timeline. 
pytrends.build_payload(keywords, timeframe="2004-01-01 2020-12-09")

# Here, we store the specific data we want. In this case, we need the popularity score or interest over time.
data = pytrends.interest_over_time()

#data.head()
#data.tail()
data
```

There are many parameters to work on like geography or location, Country, Category, timezone, search type (like image search, etc). All such parameters are set to DEFAULT. The above data is worldwide data and NOT specific to location or country.

```python
# Here we process our data to the correct format to visualize it.

data = data.reset_index()
data = data.drop(columns=["isPartial"])
#data
```

```python
# importing the necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,7))
plt.title("Popularity Score of the Keywords from 2004 to 2020", size=20)

sns.lineplot(data=data, x="date", y="Python", label="Python")
sns.lineplot(data=data, x="date", y="Java", label="Java")
sns.lineplot(data=data, x="date", y="JavaScript", label="JavaScript")
sns.lineplot(data=data, x="date", y="Ruby", label="Ruby")
sns.lineplot(data=data, x="date", y="Perl", label="Perl")

plt.xticks(size=15)
plt.yticks(size=15)
plt.ylabel("Popularity_score", size=15)
plt.xlabel("Year", size=15)
```

From the above lineplot we can conclude that Python is the most searches programming language out of 5 keywords. Java and Javascript has dropped in popularity.

The data is not 100% accurate. Comparison between certain languages like C++ and R have shown faulty graphs (unlikely numbers).
